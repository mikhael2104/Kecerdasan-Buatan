## Latihan 1 - Central Tendency

```python
#mendapatkan nilai mean
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.mean(score)
print("Mean dari deret angka tersebut adalah:", x)
```

```python
#mendapatkan nilai median
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.median(score)
print("Median dari deret angka tersebut adalah:", x)
```
