# **JOBSHEET 3 - PENGUKURAN STATISTIK DESKRIPTIF**

## Latihan 1 - Central Tendency

```python
#mendapatkan nilai mean
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.mean(score)
print("Mean dari deret angka tersebut adalah:", x)
```

```python
#mendapatkan nilai median
import numpy as np
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.median(score)
print("Median dari deret angka tersebut adalah:", x)
```

```python
#mendapatkan nilai modus/mode
from scipy import stats
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = stats.mode(score, keepdims=True)
print(x)
```

## Latihan 2 - Spread

```python
#mendapatkan nilai range
import numpy as np

data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]

minimum = np.min(data)
maximum = np.max(data)
x = maximum - minimum

print("Range dari deret data tersebut adalah:", x)
```

```python
#Mendapatkan nilai percentile
import numpy as np

data = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.percentile(data, 90)
print(x)
```

```python
#Mendapatkan nilai Quartile dan Interqurtile Range
import numpy as np

data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9 ,10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)
```

```python
#Mendapatkan nilai variance dan Standart Deviation
import numpy as np

data= [86, 87, 88, 86, 87, 85, 86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)

print(mean, var, std)
```

Tugas 3 - Pada sebuah kelas memiliki nilai hasil ujian yaitu 50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53, 63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65, 65, 53, 99. Beradasarkan data hasil ujian tersebut, tentukan :
**a.**	Urutkan data dari yang terkecil sampai data yang terbesar
**b.**	Identifikasi nilai first quartile (Q1) dan third quartile (Q3), dan IQR
**c.**	Tentukan nilai upper bound dan lower bound
**d.**	Tentukan nilai ujian yang menjadi outlier berdasarkan data hasil ujian diatas


```python
import numpy as np

data = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53,
        63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65]

#a. Mengurutkan data
sorted_data = sorted(data)

#b. Identifikasi nilai first quartile dan third quartile, dan IQR
Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1

# c. Menentukan Upper Bound dan Lower Bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# d. Identifikasi Outlier
outliers = [x for x in sorted_data if x < lower_bound or x > upper_bound]

# Output Hasil
print(f"a. Data Terurut: {sorted_data}")
print(f"b. First Quartile (Q1): {Q1}")
print(f"   Third Quartile (Q3): {Q3}")
print(f"   IQR: {IQR}")
print(f"c. Lower Bound: {lower_bound}")
print(f"   Upper Bound: {upper_bound}")
print(f"d. Nilai Outlier: {outliers}")
```

Tugas 4 - Standart Deviasi dari data berat badan mahasiswa informatika

```python
import math

# Data: (Batas Bawah, Batas Atas, Frekuensi)
data = [
    (41, 45, 2), (46, 50, 3), (50, 55, 6),
    (56, 60, 4), (61, 65, 5), (66, 70, 2)
]

# Hitung n dan jumlah fx
n = sum(f for _, _, f in data)
sum_fx = sum(((bb + ba) / 2) * f for bb, ba, f in data)
mean = sum_fx / n

# Hitung jumlah f(xi - mean)^2
sum_f_diff_sq = sum(f * (((bb + ba) / 2) - mean)**2 for bb, ba, f in data)

# Standar Deviasi (Sampel)
std_dev = math.sqrt(sum_f_diff_sq / (n - 1))

print(f"Rata-rata: {mean:.2f}")
print(f"Standar Deviasi: {std_dev:.2f}")
```
