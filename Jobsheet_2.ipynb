## Latihan 3 - Menggunakan Fungsi Math

```python
import math

# 1. Menggunakan Metode .format()
# Bagus untuk mengatur presisi desimal seperti {:.2f}
sudut = math.pi / 4  # 45 derajat
print("math.sin({:.2f})       : {}".format(sudut, math.sin(sudut)))
print("math.cos({:.2f})       : {}".format(sudut, math.cos(sudut)))
print("math.tan({:.2f})       : {}".format(sudut, math.tan(sudut)))

print("-" * 40)

# 2. Menggunakan Koma (,)
# Metode paling sederhana, Python otomatis menambahkan spasi
print("math.log(10)          :", math.log(10))
print("math.log10(100)       :", math.log10(100))

print("-" * 40)

# 3. Konversi & Jarak
print("math.degrees(3.14)    :", math.degrees(math.pi))
titik_a = (0, 0)
titik_b = (3, 4)
print("math.dist(a, b)       :", math.dist(titik_a, titik_b))

print("-" * 40)

# 4. Aljabar & Kombinatorika
n, k = 5, 2
print("math.factorial(5)     :", math.factorial(n))
print("math.perm(5, 2)       :", math.perm(n, k))
print("math.gamma(5)         :", math.gamma(n))

print("-" * 40)

# 5. Numerik Lainnya
print("math.fmod(20, 3)      :", math.fmod(20, 3))
print("math.isnan(NaN)       :", math.isnan(float('nan')))
```

```python
import math
x = math.sin
y = math.cos
z = math.tan
a = math.log
b = math.log10
c = math.degrees
d = math.dist
e = math.factorial
f = math.fmod
g = math.isnan
h = math.gamma
i = math.perm

# Eksekusi dengan nilai contoh
print("Hasil math.sin(pi/2)      : {}".format(x(math.pi / 2)))
print("Hasil math.cos(0)         : {}".format(y(0)))
print("Hasil math.tan(pi/4)      : {}".format(z(math.pi / 4)))

print("-" * 40)

print("Hasil math.log(2.71)      : {}".format(a(math.e))) # math.e adalah bilangan natural
print("Hasil math.log10(1000)    : {}".format(b(1000)))

print("-" * 40)

print("Hasil math.degrees(pi)    : {}".format(c(math.pi)))
print("Hasil math.dist((0,0),(3,4)): {}".format(d((0, 0), (3, 4))))

print("-" * 40)

print("Hasil math.factorial(5)   : {}".format(e(5)))
print("Hasil math.fmod(10, 3)    : {}".format(f(10, 3)))
print("Hasil math.isnan(NaN)     : {}".format(g(float('nan'))))
print("Hasil math.gamma(6)       : {}".format(h(6))) # Sama dengan 5! (factorial 5)
print("Hasil math.perm(10, 3)    : {}".format(i(10, 3)))
```

## Latihan 4 - Operasi Matriks dan Vektor dengan Python

```python
#membuat matriks di python
import numpy as np
a = np.array([[80, 30, 25], [50, 20, 10]])
b = np.array([[30, 60, 15], [10, 30, 30]])

hasil = a + b
print(hasil)
```

```python
#melakukan operasi dot product
import numpy as np
a = np.array([10, 20, 15])
b = np.array([16, 67, 35])

hasil = np.dot(a, b)
print(hasil)
```

**Soal no 3** - Dua Buah Vektor A dan B merupakan vektor 3D pada koordinat kartesian. Jika Vektor A = 5i + 3j + 7k dan Vektor B = 12i - 3j + k. Maka tentukan panjang masing-masing Vektor, hasil perkalian titik A.B, dan AxB.

```python
import numpy as np

a = np.array([5, 3, 7])
b = np.array([12, 3, 1])

panjangA = np.linalg.norm(a)
panjangB = np.linalg.norm(b)

hasil_dot = np.dot(a, b)
hasil_cross = np.cross(a, b)

print("Panjang Vektor A:", format(panjangA))
print("Panjang Vektor B:", format(panjangB))
print("Hasil dot product A.B: ", format(hasil_dot))
print("Hasil cross product AxB: ", format(hasil_cross))
```
